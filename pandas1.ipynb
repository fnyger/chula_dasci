## Create series from list

```python
fruits = ['orange', 'apple', 'banana', 'watermelon', 'pineapple']

fruits_series = pd.Series(fruits)
fruits_series.name = 'Fruits'

fruits_series
```

## From dict

```python
fruits_dict = {'orange': 10, 'apple': 15, 'banana': 8, 'watermelon':
3,'pineapple': 5}

fruits_series2 = pd.Series(fruits_dict)
fruits_series2.name = 'Quantity'
fruits_series2.index.name = 'Fruit'

fruits_series2
```

## Creating dataframe from dict

```python
fruits_dict2 = {
	'Fruit': ['apple', 'banana', 'orange', 'watermelon', 'pineapple'],
	'Quantity': [15, 8, 10, 3, 5],
	'Price': [0.50, 0.30, 0.80, 1.00, 1.50]
}

fruits_df = pd.DataFrame(fruits_dict2)
fruits_df
```

## Accessing data


```python
fruits_df['Fruit'] #returns series

#multiple series at once
fruits_df[['Fruit', 'Price']] #returns dataframe

#Slicing
fruits_df[:3]
fruits_df[1:4]
```

## Slicing with loc and iloc

```python
fruits_df2 = fruits_df.set_index('Fruit') #Select that the fruit column is the indexing column

fruits_df2.loc['banana']
fruits_df2.loc['banana':'orange']
fruits_df2.loc[['banana']]
fruits_df2.loc[['banana','orange'], ['Price']]
```

```python

fruits_df2.iloc[1] #series
fruits_df2.iloc[[1]] #Dataframe

fruits_df2.iloc[1:3] #Dataframe
fruits_df2.iloc[1:3,1:] #Dataframe with 2D-slicing
```

```python
fruits_df2[(fruits_df2['Quantity']>5) & (fruits_df2['Price']>0.5)] #and
fruits_df2[(fruits_df2['Quantity']>5) | (fruits_df2['Price']>0.5)] #or
```
